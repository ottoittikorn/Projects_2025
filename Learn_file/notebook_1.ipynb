```python
import pandas as pd
df = pd.read_csv('orders.csv')

df.columns
```

```python
df[["Country", "Product"]]

df.iloc[10][5]

df.iloc[10]["Country"]
```

```python
# Filtering data 

df.head()
```

```python
# Filtering data  with and 
df[(df["Category"] == "Electronics") & (df["Country"] == "USA")]
```

```python
# Filtering data  with OR condition 
df[(df["Category"] == "Electronics") | (df["Country"] == "USA")]
```

```python
#Filtering with numeric values
df[df["Quantity"] > 2]
```

```python
# Advanced filtering with string methods, filtering starting with 'A'
df[df["CustomerName"].str.startswith("A")]
# Advanced filtering with string methods, filtering end with 'a'
df[df["CustomerName"].str.endswith("a")]
```

```python
# FIlterring with multiple values in a column
df[df["Country"].isin(["USA", "Swededn", "Brazil"])]

# Reversing conditions 
df[~df["Country"].isin(["USA", "Swededn", "Brazil"])]
```

```python
# Update a role, ,manipulating data
# Use .loc instead of .iloc for label-based indexing
# Replacing the value in the 'Product' column for a specific customer "Tim" 
df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] = "Tim"
df.loc[df["CustomerName"] == "Anna Ivanova"]
```

```python
df.loc[df["Country"] == "USA", "Country"] = "United States"
df.loc[df["Country"] == "United States"]



# Changing everything in a column to capital 
df["Country"] = df["Country"].str.upper()
df["Country"] 
```

```python
df = df.drop(39)

df.tail()
```

```python
#Modify 

df.dropna()
df.fillna({"OrderID": 0}, inplace=True)
df.rename(columns={"OrderID": "Order ID"}, inplace=True)

df
```

```python
# Analysing data 
df["Country"].value_counts()
df.groupby("Country")["Price"].sum() 
```

```python
df.to_csv('modified_orders.csv', index=False)
```
